```python
for f in filenames:
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]  #this pulls molecule name
    
    
    outfile = open(f,'r')
    data = outfile.readlines() 
    outfile.close()   #this is the same as lesson 2 - opens the file
    
    for line in data:
        if 'Final Energy' in line:
            energy_line2 = line
            words2 = energy_line2.split()
            energy = float(words2[3])
            print(molecule_name, energy)  #attaches molecule name to energy
            
            
```

```python
datafile = open('energies.txt','w+')   
#'w+' will overwrite and will create if necessary; 'a+' will append to the end of existing file a

for f in filenames:
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]  
    #this pulls molecule name
    
    
    outfile = open(f,'r')
    data = outfile.readlines() 
    outfile.close()   
    #this is the same as lesson 2 - opens the file
    
    for line in data:
        if 'Final Energy' in line:
            energy_line2 = line
            words2 = energy_line2.split()
            energy = float(words2[3])
            print(molecule_name, energy)  
            #attaches molecule name to energy
            datafile.write(F'{molecule_name} \t {energy} \n')
            #from a textfile standpoint, this is a formatted string - \t = tab; \n = new line
            #F string printing is only Python 3
            # to only print all of those decimal places {energy:.3f}, 3 floating decimal places will print
            
datafile.close() 
#the file will not be written until I close the file 
```

```python
datafile = open('energiesb.txt','w+')   
#'w+' will overwrite and will create if necessary; 'a+' will append to the end of existing file a

for f in filenames:
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]  
    #this pulls molecule name
    
    
    outfile = open(f,'r')
    data = outfile.readlines() 
    outfile.close()   
    #this is the same as lesson 2 - opens the file
    
    for line in data:
        if 'Final Energy' in line:
            energy_line2 = line
            words2 = energy_line2.split()
            energy = float(words2[3])
            print(molecule_name, energy)  
            #attaches molecule name to energy
            datafile.write(F'{molecule_name} \t {energy:.3f} \n')
            #from a textfile standpoint, this is a formatted string - \t = tab; \n = new line
            #F string printing is only Python 3
            # to only print all of those decimal places {energy:.3f}, 3 floating decimal places will print
            
datafile.close() 
#the file will not be written until I close the file 
```

```python
distance_file= os.path.join("data", "distance_data_headers.csv")

distances = numpy.genfromtxt(fname=distance_file, delimiter = ',', dtype='unicode')
print(distances)
```

```python
# looks like a list of lists, but this is really a numpy array
```

```python
headers = distances[0]
print(headers)
```

```python
# extract the numerical data (everything but the header)
data = distances[1:]
print(data)
```

```python
# this give me all numbers as strings rather than floats;  need to recast
data = float(data)
```

```python
#python thinks this mean we want to take the entire matrix and turn it into a single number
#could write a loop, but numpy has a fn that can do this
```

```python
data = data.astype(numpy.float) #float and numpy.float is about the same; outputs in sci notation
print(data)
```

```python
data[:, 0] #can look at whole first column, etc
```

```python
data[0] #is whole first row - that is an array in itself
```

```python
data [0,0] #always [row,column]
```

```python
#small_data = data[startrow:endrow, startcolumn:endcolumn]

small_data = data[0:10, 0:3]
print(small_data)
```

```python
small_data2 = data [:, 0] #all teh rows, only the first column
print(small_data2)
```

```python
#data_average = numpy.mean(data_set)

thr4_atp = data[:,1]  #slice of every row and the 2nd column
print(thr4_atp)
```

```python
avg_thr4_atp =  numpy.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
#we want average of each column.  if we don't know how many columns, need to determine number of columns

num_columns = len(data[0,:]) #could have used any row label
print(num_columns)
```

```python
for i in range(1, num_columns):   
    column = data [:, i]  #taking a slice of all rows, but now doing for each column
    avg_col = numpy.mean(column)
    print(avg_col)
```

```python
for i in range(1, num_columns):   
    column = data [:, i]  #taking a slice of all rows, but now doing for each column
    avg_col = numpy.mean(column)
    print(F'{headers[i]} {avg_col:.3f}')
```
